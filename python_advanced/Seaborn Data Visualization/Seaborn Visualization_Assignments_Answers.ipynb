**Section 1 — Relational & Categorical Plots**

**Task 1:** Import seaborn and load abalone.csv. Create a lineplot with Length on the xaxis and Diameter on the y-axis.

```python
import seaborn as sns                            # import seaborn
import matplotlib.pyplot as plt                  # import matplotlib
import pandas as pd                              # import pandas for using local data

data = pd.read_csv("../data/abalone.csv")        # load abalone
```

```python
print(data.head())                              # first look
```

```python
sns.lineplot(x='Length', y='Diameter', data=data)           # creating lineplot
```

**Task 2:** Create a scatterplot with Whole weight on the x-axis and Shell weight on the yaxis.


```python
sns.scatterplot(x='Whole weight', y='Shell weight', data=data)
plt.show()
```

**Task 3:** Create a barplot showing the mean number of Rings for each Sex category.


```python
sns.barplot(x='Sex', y='Rings', data=data)              # mean is standard in barplots
plt.show()
```

**Task 4:** Create a countplot showing how many rows exist for each Sex category.

```python
sns.countplot(x='Sex', data=data)
plt.show()
```

**Task 5:** Create a boxplot comparing Diameter across the three Sex categories.

```python
sns.boxplot(x='Sex', y='Diameter', data=data)
plt.show()
```

**Section 2 — Distribution Plots**

**Task 6:** Create a histplot of the Height column.

```python
plt.xlim(0, 0.3)                            # customizing x-axis

sns.histplot(x='Height', data=data)
plt.show()
```

**Task 7:** Create a pairplot using the Length, Diameter, Height, and Rings columns, with
hue set to Sex.

```python
sns.pairplot(data[["Length", "Diameter", "Height", "Rings", "Sex"]], hue="Sex")
plt.show()
```

**Section 3 — Multi-Plot Layouts & Styling**

**Task 8:** Using plt.subplot(), place a scatterplot of Length vs Whole weight side by side
with a boxplot of Whole weight by Sex, in one figure.

```python
def graph_1():	
	sns.scatterplot(x='Length', y='Whole weight', data=data)

def graph_2():
	sns.boxplot(x='Sex', y='Whole weight', data=data)				# sex on x-axis, because this looks more familiar to me

# Adding the scatterplot at the specified
# grid position
plt.subplot(121)
graph_1()

# Adding the boxplot at the specified
# grid position
plt.subplot(122)
graph_2()

plt.show()
```

**Task 9:** Use sns.FacetGrid() to create one panel per Sex category, each showing a
histogram of Rings.

```python
g = sns.FacetGrid(data, col="Sex")      # panels for Sex       
g.map(sns.histplot, "Rings")            # from ChatGPT

plt.show()
```

**Task 10:** Set the Seaborn style to 'whitegrid', then create a lineplot of Length vs Height
and remove the top/right border with sns.despine().


```python
# without styling for comparison

sns.lineplot(data=data, x="Length", y="Height")
plt.show()      
```

```python
sns.set_style("whitegrid")                           # setting the style

sns.lineplot(data=data, x="Length", y="Height")      # creating the lineplot

sns.despine()                                        # removing the borders

plt.show()
```
