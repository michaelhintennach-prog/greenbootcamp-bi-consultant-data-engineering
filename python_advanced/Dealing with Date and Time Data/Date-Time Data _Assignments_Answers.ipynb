**Section 3 — Timedelta Calculations**

**Task 10:** Create two datetime objects: January 1st, 2023 at 8:00 AM, and June 15th,
2024 at 5:30 PM. Subtract the earlier one from the later one and print the resulting
timedelta, along with its .days and .seconds attributes.

```python
dt1 = datetime(2023, 1, 1, 8, 0)
print(dt1)
dt2 = datetime(2024, 6, 15, 17, 30)
print(dt2)

td1 = dt2 - dt1
print(td1)
print(td1.days, "days")
print(td1.seconds, "seconds")
```

**Task 11:** Take that same duration and convert it into a number of hours, and separately
into a number of minutes

```python
# no attributes for hours and minutes in timedelta -> use the days and seconds
# td1 = 531 days, 34200 seconds

print("Hours: ", td1.days * 24 + td1.seconds / 3600)
print("Minutes: ", td1.days * 24 * 60 + td1.seconds / 60)
```

**Task 16:** Build a small DataFrame with three columns: Order_date (8 consecutive days
starting 2024-01-01), Ship_date (8 weekly dates starting 2024-01-01), and Priority (8
random integers from 1 to 3). Then add two more columns using the .dt accessor:
Order_month (the month number of Order_date) and Ship_day_of_week (the day name
of Ship_date).

```python
import numpy as np       # import numpy for random numbers

order_dates = pd.date_range(start="2024-01-01", periods=8, freq="D")        # 8 consecutive days
ship_dates = pd.date_range(start="2024-01-01", periods=8, freq="7D")        # 8 weekly dates - freq "W" would return 2024-01-07 in the first row

priority = np.random.randint(1, 4, size=8)                                  # 8 random integers from 1 to 3

# Building the DataFrame
df = pd.DataFrame({"Order_date": order_dates, "Ship_date": ship_dates, "Priority": priority})
df
```

```python
df["Order_month"] = df["Order_date"].dt.month               # adding order month
df["Ship_day_of_week"] = df["Ship_date"].dt.day_name()      # adding ship day
df
```
