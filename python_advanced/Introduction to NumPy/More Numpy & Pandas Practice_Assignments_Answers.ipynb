**Section 2 — Combining, Binning, and Pivoting**

**Task 10:** Read both wine datasets. Find the mean sulphates for red and white
separately. For each type, build a flag column marking which wines are ABOVE their
own type's mean sulphates, attach it using both .join() and pd.concat(), then return just
the rows that are AT OR BELOW the mean.


```python
# import pandas
import pandas as pd

# load the red wine data
redwine = pd.read_csv('../data/winequality-red.csv', delimiter=';')

# load the white wine data
whitewine = pd.read_csv('../data/winequality-white.csv', delimiter=';')
```

```python
# getting the means 

red_mean = (redwine["sulphates"].mean())
white_mean = (whitewine["sulphates"].mean())
print(red_mean)
print(white_mean)
```

```python
# building the flag columns

red_flag = (redwine["sulphates"] > red_mean).rename("sulphates_above_mean")
white_flag = (whitewine["sulphates"] > white_mean).rename("sulphates_above_mean")
```

```python
# attaching flag columns using join

redwine_joined = redwine.join(red_flag)
whitewine_joined = whitewine.join(white_flag)

whitewine_joined.head()
```

```python
# attaching flag columns using concat

redwine_concat = pd.concat([redwine, red_flag], axis=1)
whitewine_concat = pd.concat([whitewine, white_flag], axis=1)

redwine_concat.head()
```

```python
# returning low sulphates wines

red_low_sulphates = redwine_joined[redwine_joined["sulphates_above_mean"] == False]
white_low_sulphates = whitewine_joined[whitewine_joined["sulphates_above_mean"] == False]
```

**Task 11:** Bin red wine's pH column into 4 equal-width bins. Attach the bin labels to the
DataFrame, then return every row that is NOT in the highest pH bin.

```python
redwine_bins = pd.cut(redwine["pH"], bins=4)                                            # creating 4 bins for pH values
redwine_bins_concat = pd.concat([redwine, redwine_bins.rename("pH_bins")], axis=1)      # adding the renamed bins as a column to the original dataframe
top_bin = redwine_bins.cat.categories[-1]                                               # computing the top bin
redwine_not_top = redwine_bins_concat[redwine_bins_concat["pH_bins"] != top_bin]        # filtering out the top bin

print(redwine_not_top.shape[0], "of", redwine.shape[0], "red wines have pH values that are not in the top bin.")
redwine_not_top
```

**Task 13:** Read bike_share_data.csv and clean its column names (lowercase,
underscores). Create a duration_group column using pd.cut() with these bin edges: 0,
300, 600, 1200, 3600, 300000 seconds, labeled '0-5min', '5-10min', '10-20min', '20-
60min', '60min+'. Build a pivot table with average duration, using BOTH
subscription_type and duration_group together as the index.

```python
# load bike_share data

bike_share = pd.read_csv('../data/bike_share_data.csv')
bike_share.head()
```

```python
# create duration bins and labels

bins = [0, 300, 600, 1200, 3600, 300000]
labels = ['0-5 min', '5-10 min', '10-20 min', '20-60 min', '60+ min']

bike_share['duration_bins'] = pd.cut(bike_share["Duration"], bins=bins, labels=labels)
```

```python
# pivot table for duration bins

bike_share_pivot = bike_share.pivot_table(index=["Subscription Type", "duration_bins"], values="Duration", aggfunc="mean")
bike_share_pivot
```
