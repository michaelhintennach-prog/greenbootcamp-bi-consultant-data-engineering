**Section 3 — Indexing & Aggregation**

**Task 7:** Build a 4x5 array using arange() and reshape(), containing the numbers 0
through 19. Grab row 1 on its own, then column 3 on its own, then the sub-block
spanning rows 1–2 and columns 2–3.

```python
range_array = np.arange(0, 20, 1).reshape(4, 5)         # end-argument = 20
range_array
```

```python
row1 = range_array[1]                   # no second index -> grabbing the row
print(row1)
```

```python
column3 = range_array[:, 3]             # column 3 of each row
print(column3)
```

```python
subset = range_array[0:2, 1:3]          # column 2 and 3 of rows 1 and 2, indices are end exclusive
print(subset)
```

**Task 12:** Read winequality-white.csv. Use pd.cut() to split the alcohol column into 5
equal-width bins, name the new column alcohol_bin, and attach it to the DataFrame.
Then build a pivot table showing average residual sugar, with quality as the rows and
alcohol_bin as the columns.

```python
import pandas as pd                 # importing pandas for reading csv
```

```python

white_wine_df = pd.read_csv('../data/winequality-white.csv', delimiter=';')     # loading white wine data
white_wine_df.shape
white_wine_df.columns
```

```python
white_wine_df["alcohol_bin"] = pd.cut(white_wine_df["alcohol"], bins=5)                     # splitting alcohol column into 5 bins
```

```python
pd.pivot_table(white_wine_df, values="residual sugar", index="quality", columns="alcohol_bin", aggfunc="mean")
```
